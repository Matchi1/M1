```python
# Exercice 1.A
def is_palindrome(word):
    return word[::-1] == word

print(is_palindrome("hello"))
print(is_palindrome("abbaabba"))
```

```python
# Exercice 1.B
import string
def is_pangram(word):
    return not -1 in [word.strip().lower().find(l) for l in string.ascii_lowercase]
print(is_pangram("The quick brown fox jumps over the lazy dog"))
print(is_pangram("Portez ce vieux whisky au juge blond qui fume."))
print(is_pangram("hello"))
```

```python
def remove_adjacent(lst):
    return [lst[i - 1] for i in range(1, len(lst) + 1) if (i == len(lst) or lst[i - 1] != lst[i])]
print(remove_adjacent([1,2,2,3,3,1]))
```

```python
def digits_sum(number):
    return sum([int(i) for i in list(str(number))])
print(digits_sum(555))
print(digits_sum(51545))
```

```python
# Exercice 2
def read_file(filename):
    lines = []
    with open(filename, "r") as file:
        lines = file.readlines()
    return lines

def word_follow_rule(rule, word):
    l = list(word.strip('\n'))
    for letter in rule:
        for i in range(len(l)):
            if letter == l[i]:
                l.pop(i)
                break
    if l == []:
        return True
    return False

def words_follow_rule(rule, dico):
    words = []
    for w in dico:
        if word_follow_rule(rule, w.upper()):
            words.append(w.upper())
    return words

def determine_max_len(words):
    max_len = 0
    for w in words:
        max_len = max(len(w), max_len)
    return max_len

def find_word_by_len(length, words):
    result = []
    for w in words:
        if len(w) == length:
            result.append(w)
    return result

def lookup(rule):
    lines = read_file('/usr/share/dict/french')
    words = words_follow_rule(rule, lines)
    max_len = determine_max_len(words)
    return find_word_by_len(max_len, words)
```

```python
# Exercice 3

import random
import re

def get_rand_number(a, b):
    return random.randint(a, b)

def shuffle_word(word):
    l = list(word)
    size = len(word) - 2
    for i in range(1, len(word) - 1):
        rand_index = get_rand_number(1, size)
        l[i], l[rand_index] = l[rand_index], l[i]
    return ''.join(l)

def read_file(filename):
    lines = []
    with open(filename, "r", encoding='latin-1') as file:
        lines = file.readlines()
    return lines

def reverse(filename):
    lines = read_file(filename)
    text = ""
    for line in lines:
        words = re.split("(\W)", line)
        for i in range(len(words)):
            if not words[i].isspace():
                words[i] = shuffle_word(words[i])
        text += ''.join(words)
    return text
    
print(reverse("harmonie.txt"))
```
