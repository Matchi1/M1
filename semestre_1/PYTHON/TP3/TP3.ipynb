```python
# 1.2
def list_dir(dirname):
    return list(os.walk(dirname))

def count(dirname):
    l = list_dir(dirname)
    count = 0
    for directory in l:
        _, _, filenames = directory
        count += len(filenames)
    return len(l), count

def count_file_and_dir(dirname):
    nb_dir, nb_file = count(dirname)
    print("directory: " + dirname)
    print("number of file : {}".format(nb_file))
    print("number of directory : {}".format(nb_dir))
```

```python
def count(dirname):
    l = list_dir(dirname)
    count_file, count_link = 0, 0
    count_public, count_private = 0, 0
    for directory in l:
        (dirname, dirnames, filenames) = directory
        for f in filenames:
            path = os.path.join(dirname, f)
            if os.path.islink(path):
                count_link += 1
            else:
                count_file += 1
            if os.stat(dirname).st_uid < os.getuid():
                count_private += 1
            else:
                count_public += 1
    return len(l), count_link, count_file, count_private, count_public

def count_file_and_dir_and_link_and_stat(dirname):
    nb_dir, nb_link, nb_file, private, public = count(dirname)
    print("directory: " + dirname)
    print("number of file : {}".format(nb_file))
    print("number of directory : {}".format(nb_dir))
    print("number of link : {}".format(nb_link))
    print("number of private file : {}".format(private))
    print("number of public file : {}".format(public))
        
```

```python
import re
def plusOneGroup(word):
    index = 0
    result = []
    for pos in range(len(word)):
        l = []
        for index in range(len(word)):
            if index == pos:
                l.append('.')
            l.append(word[index])
            
        result.append(''.join(l))
    l = list(word)
    l.append('.')
    result.append(''.join(l))
    return result

def minusOneGroup(word):
    result = []
    for i in range(len(word)):
        l = []
        for index in range(len(word)):
            if i != index:
                l.append(word[index])
        result.append(''.join(l))
    return result

def equalZeroGroup(word):
    result = []
    for i in range(len(word)):
        l = []
        for index in range(len(word)):
            if i == index:
                l.append('.')
            else:
                l.append(word[index])
        result.append(''.join(l))
    return result

def matchDico(pattern):
    result = []
    lines = open('/usr/share/dict/french', encoding='latin1').readlines()
    for line in lines:
        w = line.strip('\n')
        if re.fullmatch(pattern, w) is not None:
            result.append(w)
    return result

def lookup(word):
    pattern = []
    pattern.extend(plusOneGroup(word))
    pattern.extend(minusOneGroup(word))
    pattern.extend(equalZeroGroup(word))
    regex = '(' + '|'.join(pattern) + ')'
    
    return matchDico(regex)
    
print(lookup('ais'))
```
