```python
def tabadd(n):
    print(' + |', end=' ')
    for y in range(n): 
        print ("%2d "% (y % n),end=' ') 
    print()
    print('-'*(4*n+3))
    for x in range(n): 
        print('%2d |'%x, end= ' ')
        for y in range(n): 
            print ("%2d "% ((x+y) % n),end=' ') 
        print()
```

```python
tabadd(26)
```

```python
def tabmul(n):
    print(' * |', end=' ')
    for y in range(1,n): 
        print ("%2d "% (y % n),end=' ') 
    print()
    print('-'*(4*n-1))
    for x in range(1,n): 
        print('%2d |'%x, end= ' ')
        for y in range(1,n): 
            print ("%2d "% ((x*y) % n),end=' ') 
        print()
```

```python
tabmul(26)
```

```python
def msg_size(message):
    size = 0
    for i in message:
        if i in AA:
            size += 1
    return size

def coincidence(message):
    result = 0
    oo = occurence(message)
    size = msg_size(message)
    for letter,occ in oo:
        if letter in AA:
            result += (occ * (occ - 1))/(size * (size - 1))
    return result

msg = ''
with open("mono.txt") as f:
    msg = f.read()
print(coincidence(msg))
with open("poly.txt") as f:
    msg = f.read()
print(coincidence(msg))
```

```python
def length_key(message):
    result = 0
    for i in range(1,500):
        size = msg_size(message[::i])
        index = coincidence(message[::i])
        # french
        index_A = 1/26
        index_L = 0.074
        k = (size * (index_L - index_A))/((size - 1) * index - size * index_A + index_L)
        if result < k:
            result = k
        print(result)
    return result

with open("mono.txt") as f:
    msg = f.read()
print(length_key(msg))
```
